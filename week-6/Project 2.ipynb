```python
import tkinter as tk
from tkinter import messagebox

# Create main window
root = tk.Tk()
root.title("Admission Validation")
root.geometry("500x200")

# Create labels and entries for name & Course
name_label = tk.Label(root,text="Input your Name")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Define options for the drop-down menu
options = ["Computer Science", "Mass Communication"]

# Create a variable to store the selected option
course_selected = tk.StringVar(root)
course_selected.set("Select")

course_label = tk.Label(root,text="Choose your course")
course_label.pack()
course_entry = tk.OptionMenu(root, course_selected, *options)
course_entry.pack()

# Submit Button Function
def submit():
    name = name_entry.get()
    course = course_selected.get().lower()

    if len(name) == 0 or course not in ["computer science", "mass communication"]:
        if len(name) == 0 and course not in ["computer science", "mass communication"]:
            messagebox.showinfo("Validation Failed", "Please input name and select a course")
        elif len(name) == 0:
            messagebox.showerror("Validation Failed", "Please input name")
        else:
            messagebox.showerror("Validation Failed","Please select a course")
        return
    
    messagebox.showinfo("Validation Success",f"Welcome {name}!")
    # Create computer science window
    if course == "computer science":
        window_1 = tk.Tk()
        window_1.title("Computer Science Portal")
        window_1.geometry("1000x500")

        # Create label and menu for JAMB score
        jamb_label = tk.Label(window_1, text="Input your JAMB Score")
        jamb_label.pack()
        jamb_entry = tk.Entry(window_1)
        jamb_entry.pack()

        # Create array to store grade options
        grade_options= ["A1","B2","B3","C4","C5","C6","D7","E8","F9"]

        # Create variables to store grade selections
        math_grade = tk.StringVar(window_1)
        math_grade.set("Select")

        eng_grade = tk.StringVar(window_1)
        eng_grade.set("Select")

        phy_grade = tk.StringVar(window_1)
        phy_grade.set("Select")

        chem_grade = tk.StringVar(window_1)
        chem_grade.set("Select")

        fmath_grade =tk.StringVar(window_1)
        fmath_grade.set("Select")

        # Create labels and entries(Option Menu) for grades 
        math_grade_label = tk.Label(window_1, text="Math grade")
        math_grade_label.pack()
        math_grade_entry = tk.OptionMenu(window_1, math_grade, *grade_options)
        math_grade_entry.pack()

        eng_grade_label = tk.Label(window_1, text="English grade")
        eng_grade_label.pack()
        eng_grade_entry = tk.OptionMenu(window_1, eng_grade, *grade_options)
        eng_grade_entry.pack()

        phy_grade_label = tk.Label(window_1, text="Physics grade")
        phy_grade_label.pack()
        phy_grade_entry = tk.OptionMenu(window_1, phy_grade, *grade_options)
        phy_grade_entry.pack()

        chem_grade_label = tk.Label(window_1, text="Chemistry grade")
        chem_grade_label.pack()
        chem_grade_entry = tk.OptionMenu(window_1, chem_grade, *grade_options)
        chem_grade_entry.pack()

        fmath_grade_label = tk.Label(window_1, text="Further Math grade")
        fmath_grade_label.pack()
        fmath_grade_entry = tk.OptionMenu(window_1, fmath_grade, *grade_options)
        fmath_grade_entry.pack()

        # Create variable to store the yes/no response
        yesno_response = tk.IntVar(window_1)

        # Create label and entry(radio button) for interveiew
        interview_label = tk.Label(window_1, text="Did you pass your interview?")
        interview_label.pack()
        yes_button = tk.Radiobutton(window_1, text="Yes", variable=yesno_response, value=1)
        yes_button.pack()
        no_button = tk.Radiobutton(window_1, text="No", variable=yesno_response, value=0)
        no_button.pack()

        admitted_students = []
        not_admitted_Students = []
        
        def submit_comp():
            # Validate input
            jamb = int(jamb_entry.get())
            math = math_grade.get()
            eng = eng_grade.get()
            phy = phy_grade.get()
            chem = chem_grade.get()
            fmath = fmath_grade.get()
            interview = yesno_response.get()

            student_info = {
                "Name": name,
                "Course": course,
                "Jamb Score": jamb,
                "Math Grade": math,
                "English Grade": eng,
                "Physics Grade": phy,
                "Chemistry Grade": chem,
                "Further Math Grade": fmath,
                "Interview Result": "Pass" if interview == 1 else "Fail"
            }

            if jamb > 400:
                messagebox.showerror("Admission Status","Please enter a valid JAMB score")
            
            elif jamb <= 400 and jamb >= 250:
                if interview == 1:
                    grades_required = ['A1','B2','B3','C4','C5','C6']
                    if math in grades_required and eng in grades_required and phy in grades_required and chem in grades_required and fmath in grades_required:
                        messagebox.showinfo("Admission Status", "Congratulations! You have been admitted")
                        admitted_students.append(student_info)
                    else:
                        messagebox.showerror("Admission Status", "Sorry, your grades do not meet the requirements")
                        not_admitted_students.append(student_info)
                else:
                    messagebox.showerror("Admission Status", "Sorry, you did not pass the interview")
                    not_admitted_students.append(student_info)
            else:
                messagebox.showerror("Admission Status", "Sorry, your JAMB score does not meet the requirements")
                not_admitted_students.append(student_info)

            print(admitted_students)
                    
        submit_comp_button = tk.Button(window_1, text="Submit", command=submit_comp)
        submit_comp_button.pack()

    if course == "mass communication":
            window_2 = tk.Tk()
            window_2.title("Mass Communication Portal")
            window_2.geometry("1000x500")
    
            # Create label and menu for JAMB score
            mc_jamb_label = tk.Label(window_2, text="Input your JAMB Score")
            mc_jamb_label.pack()
            mc_jamb_entry = tk.Entry(window_2)
            mc_jamb_entry.pack()
    
            # Create array to store grade options
            mc_grade_options= ["A1","B2","B3","C4","C5","C6","D7","E8","F9"]
    
            # Create variables to store grade selections
            mc_math_grade = tk.StringVar(window_2)
            mc_math_grade.set("Select")
    
            mc_eng_grade = tk.StringVar(window_2)
            mc_eng_grade.set("Select")
    
            gov_grade = tk.StringVar(window_2)
            gov_grade.set("Select")
    
            crs_grade = tk.StringVar(window_2)
            crs_grade.set("Select")
    
            econs_grade =tk.StringVar(window_2)
            econs_grade.set("Select")
    
            # Create labels and entries(Option Menu) for grades 
            mc_math_grade_label = tk.Label(window_2, text="Math grade")
            mc_math_grade_label.pack()
            mc_math_grade_entry = tk.OptionMenu(window_2, mc_math_grade, *mc_grade_options)
            mc_math_grade_entry.pack()
    
            mc_eng_grade_label = tk.Label(window_2, text="English grade")
            mc_eng_grade_label.pack()
            mc_eng_grade_entry = tk.OptionMenu(window_2, mc_eng_grade, *mc_grade_options)
            mc_eng_grade_entry.pack()
    
            gov_grade_label = tk.Label(window_2, text="Government grade")
            gov_grade_label.pack()
            gov_grade_entry = tk.OptionMenu(window_2, gov_grade, *mc_grade_options)
            gov_grade_entry.pack()
    
            crs_grade_label = tk.Label(window_2, text="CRS grade")
            crs_grade_label.pack()
            crs_grade_entry = tk.OptionMenu(window_2, crs_grade, *mc_grade_options)
            crs_grade_entry.pack()
    
            econs_grade_label = tk.Label(window_2, text="Economics grade")
            econs_grade_label.pack()
            econs_grade_entry = tk.OptionMenu(window_2, econs_grade, *mc_grade_options)
            econs_grade_entry.pack()
    
            # Create variable to store the yes/no response
            mc_yesno_response = tk.IntVar(window_2)
    
            # Create label and entry(radio button) for interveiew
            mc_interview_label = tk.Label(window_2, text="Did you pass your interview?")
            mc_interview_label.pack()
            mc_yes_button = tk.Radiobutton(window_2, text="Yes", variable=mc_yesno_response, value=1)
            mc_yes_button.pack()
            mc_no_button = tk.Radiobutton(window_2, text="No", variable=mc_yesno_response, value=0)
            mc_no_button.pack()
    
            mc_admitted_students = []
            mc_not_admitted_Students = []
            
            def submit_mc():
                # Validate input
                mc_jamb = int(mc_jamb_entry.get())
                mc_math = mc_math_grade.get()
                mc_eng = mc_eng_grade.get()
                gov = gov_grade.get()
                crs = crs_grade.get()
                econs = econs_grade.get()
                mc_interview = mc_yesno_response.get()
    
                mc_student_info = {
                    "Name": name,
                    "Course": course,
                    "Jamb Score": mc_jamb,
                    "Math Grade": mc_math,
                    "English Grade": mc_eng,
                    "Government Grade": gov,
                    "CRS Grade": crs,
                    "Economics Grade": econs,
                    "Interview Result": "Pass" if mc_interview == 1 else "Fail"
                }
    
                if mc_jamb > 400:
                    messagebox.showerror("Admission Status","Please enter a valid JAMB score")
                
                elif mc_jamb <= 400 and mc_jamb >= 250:
                    if mc_interview == 1:
                        mc_grades_required = ['A1','B2','B3','C4','C5','C6']
                        if mc_math in mc_grades_required and mc_eng in mc_grades_required and gov in mc_grades_required and crs in mc_grades_required and econs in mc_grades_required:
                            messagebox.showinfo("Admission Status", "Congratulations! You have been admitted")
                            mc_admitted_students.append(mc_student_info)
                        else:
                            messagebox.showerror("Admission Status", "Sorry, your grades do not meet the requirements")
                            mc_not_admitted_students.append(mc_student_info)
                    else:
                        messagebox.showerror("Admission Status", "Sorry, you did not pass the interview")
                        mc_not_admitted_students.append(mc_student_info)
                else:
                    messagebox.showerror("Admission Status", "Sorry, your JAMB score does not meet the requirements")
                    mc_not_admitted_students.append(mc_student_info)
    
                print(mc_admitted_students)
                
            submit_mc_button = tk.Button(window_2, text="Submit", command=submit_mc)
            submit_mc_button.pack()

# Create Submit Button
submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
