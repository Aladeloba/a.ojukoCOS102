```python
import tkinter as tk
from tkinter import messagebox

# Create main window
root = tk.Tk()
root.title("Delivery Screen")
root.geometry("500x200")

# Create location label and entry
location_label = tk.Label(root, text="Location:")
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

# Create weight label and entry
weight_label = tk.Label(root, text="Weight in kg:")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

# Function to return cost based on location and weight
def calculate_cost(location,weight):
    if location == "ibeju-lekki":
        if weight < 10.0:
            return "₦3,500"
        else:
            return "₦5,000"
    elif location == "epe":
        if weight < 10.0:
            return "₦5,000"
        else:
            return "₦10,000"
    else:
        return "We do not currently deliver to your location."
        
# Submit button function
def submit():
    location = location_entry.get().lower()
    weight_str = weight_entry.get()

# A try- except block to catch if weight input is wrong datatype
    try:
        weight = float(weight_str)
        cost = calculate_cost(location,weight)
        messagebox.showinfo("Payment Screen", f"Total cost is {cost}")
    except:
        messagebox.showerror("Error", "Invalid weight, Please enter a valid number.")

# Create Submit button
submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```

```python
import tkinter as tk

def on_selection_change(event):
    selected_option = variable.get()
    if selected_option == "Select":
        print("Please select an option.")
    else:
        print("Selected option:", selected_option)

# Create main window
root = tk.Tk()
root.title("Drop-down Entry Menu")
root.geometry("300x200")

# Define options for the drop-down menu
options = ["Select", "Option 1", "Option 2", "Option 3", "Option 4"]

# Create a variable to hold the selected option
variable = tk.StringVar(root)
variable.set(options[0])  # Set the default option

# Create the drop-down menu
dropdown_menu = tk.OptionMenu(root, variable, *options, command=on_selection_change)
dropdown_menu.pack()

root.mainloop()
```
